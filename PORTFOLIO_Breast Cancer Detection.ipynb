```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

```python
#test and train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
def models(X_train,Y_train):
    #Logistic  Regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(X_train,Y_train)
    #Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(X_train,Y_train)
    #Print model accuracy
    print('[0]Logistic Regression Traning Score: ',log.score(X_train,Y_train))
    print('[1]Decision Tree Classifier Traning Score: ',tree.score(X_train,Y_train))
    print('[2]Random Forest Classifier Traning Score: ',forest.score(X_train,Y_train))
    
    return log,tree,forest
```

```python
#giving all of models
model=models(X_train,y_train)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
for i in range(len(model)):
    print('Model: ',i)
    cm=confusion_matrix(y_test,model[i].predict(X_test))
    print(cm)
    TP=cm[0][0]
    TN=cm[1][1]
    FN=cm[1][0]
    FP=cm[0][1]
    print('Testing Accuracy: ',(TP+TN)/(TP+TN+FN+FP))
```

```python
from sklearn.metrics import classification_report,accuracy_score
for i in range(len(model)):
    print('Model: ',i)
    print(classification_report(y_test,model[i].predict(X_test)))
    print(accuracy_score(y_test,model[i].predict(X_test)))
    print()
```
