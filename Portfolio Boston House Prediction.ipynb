```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston
```

```python
dataset=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
dataset['target']=boston.target
```

```python
dataset
```

```python
dataset.describe()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(dataset['CRIM'])
```

```python
min_treshold=dataset['CRIM'].quantile(0.05)
```

```python
max_treshold=dataset['CRIM'].quantile(0.95)
```

```python
dataset=dataset[(dataset["CRIM"]>min_treshold) & (dataset["CRIM"]<max_treshold)]
```

```python
dataset
```

```python
X=dataset.drop(['target'],axis=1)
```

```python
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_test,y_test)
```

```python
y_predict=reg.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
#check model accuracy
```

```python
print(np.mean(y_predict-y_test)**2)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test,y_predict))
```
